### 1. Exploratory analysis

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('boat_data.csv')
```

```python
data.info()
```

```python
data.head(6)
```

You can see there are currency and amounts in the Price column. We will standardize the currency to USD. Need to split the currency and numbers, then convert to USD.

1 CHF = 1.09 USD

1 EUR = 1.13 USD

1 DKK = 0.15 USD

1 Â£ = 1.35 USD

1 SEK = 0.096 USD

```python
data['Price'] = data['Price'].astype(str)
data['Price'] = data['Price'].str.replace(".", '')

data['Currency'] = data['Price'].str[:3]

data['Amount'] = data['Price'].str[4:10].str.strip()
data['Amount'] = data['Amount'].str.replace(",", '')
data['Amount'] = data['Amount'].str.replace("-", '')
data['Amount'] = pd.to_numeric(data['Amount'], errors='coerce')

# Replace NaN values with a default value (e.g., 0)
data['Amount'].fillna(0, inplace=True)
data['Amount'] = data['Amount'].astype(int)

# Currency values to USD
dict = {'CHF' : 1.09, 'EUR' : 1.13, 'DKK': 0.15, 'Â£ ' : 1.35, 'SEK' : 0.096}
data.replace({'Currency': dict},inplace=True)

# Form USD price
data['USD'] = data['Currency'] * data['Amount']
data['USD'] = pd.to_numeric(data['USD'], errors='coerce')
```

```python
pd.options.display.float_format = '{:.2f}'.format
data['USD'].describe()
```

```python
data = data.drop(['Price', 'Location', 'Currency','Amount'], axis=1)
```

### 2. Modeling

#### 2.1 Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.compose import TransformedTargetRegressor
```

```python
X = data.drop(['USD'], axis=1)
y = data['USD']

X = sc.fit_transform(X)
y = np.log10(data['USD'])
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=29)

model = LinearRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)

mse = mean_squared_error(y_test, pred)
print(mse)
```

```python
r2_score(y_test, pred)
```

Actual vs. predicted data plotting:

```python
plt.scatter(y_test, pred, color='blue')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```
